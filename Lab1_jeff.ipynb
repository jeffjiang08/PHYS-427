```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

# generate many data point for normal distribution with mu and sigma
mu, sigma = 0, 1
norm_rand = np.random.normal(mu, sigma, 10000)

# generate many data point for exponential distribution with lambda
lamd = 1
exp_rand = np.random.exponential(lamd,10000)

# graph the histogram for the normal distribution
fig, ax = plt.subplots(1, 1)
ax.hist(norm_rand,50, density=True)
x = np.linspace(-3,3,1000)
ax.plot(x,stats.norm.pdf(x,mu, sigma),linewidth = 3)
plt.title("Normal Distribution")
plt.xlabel('Values')
plt.ylabel('Probability Density')

# graph the histogram for the exponertial distribution
fig, ax = plt.subplots(1, 1)
ax.hist(exp_rand,50, density=True)
x = np.linspace(0,6,1000)
ax.plot(x,stats.expon.pdf(x),linewidth = 3)
plt.title("Exponertial Distribution")
plt.xlabel('Values')
plt.ylabel('Probability Density')

# graph the scatter plots of the normal distributions
fig, ax = plt.subplots(1, 1)
norm_rand2 = np.random.normal(mu, sigma, 10000)
ax.plot(norm_rand,norm_rand2,'.')
plt.title("Scatter Plots of Normal Distribution")

# graph the scatter plots of the normal distributions
fig, ax = plt.subplots(1, 1)
exp_rand2 = np.random.exponential(lamd,10000)
ax.plot(exp_rand,exp_rand2,'.')
plt.title("Normal Distribution")
```
